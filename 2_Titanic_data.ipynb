```python
import random
import numpy as np
import matplotlib as plt
import seaborn as sns
import pandas as pd

random.seed(23)
N=100
a=[x for x in range(N)]
b =[random.gauss(155,20) for x in range (N)]
c =[random.gauss(100,15) for x in range(N)]
d =[random.gauss(0,3) for x in range(N)]

thedata =pd.DataFrame({"A":a,"B":b,"C":c, "Category":d})
```

```python
sns.scatterplot(x="A",y="B",data=thedata)
```

```python
sns.scatterplot(x="A",y="B",hue='Category',data=thedata)
```

```python
p = sns.countplot(data=thedata,x='Category')
sns.barplot(data=thedata,x="A",y="B")
```

```python
sns.lineplot(data=thedata.drop(["Category"],axis=1))
```

```python
sns.displot(thedata['C'], bins =10, kde=False)
```

```python
sns.displot(thedata['C'], bins =10, kde=True)
```
